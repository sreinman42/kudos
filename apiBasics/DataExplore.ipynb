```python
import json
import pandas as pd

file_name = 'luke.json'
raw_Df = pd.read_json(file_name)
raw_Df
```

```python
cols = ['distance','moving_time','total_elevation_gain','total_photo_count','average_speed','max_speed','achievement_count','kudos_count','average_heartrate','len_title']
data = []

f = open(file_name)
activities = json.load(f)
f.close()

for activity in activities:                        
        lst = []
        for col in cols[:-2]: 
            lst.append(activity[col])
        if activity['has_heartrate']:
            lst.append(activity['average_heartrate'])
        else:
            lst.append(-1)
        lst.append(len(activity['name']))
        data.append(lst)
        
        
df = pd.DataFrame(data=data, columns=cols)
df
```
